```python
import numpy as np
mean = [-3, 0]
cov = ([[1, 0.8], [0.8, 1]])
np.random.seed(0)
random_numbers= np.random.multivariate_normal(mean, cov, 500)
print(random_numbers.ndim)
print(random_numbers.shape)
print(random_numbers)
```

```python
import numpy as np
mean = [-3, 0]
cov = [[1, 0.8], [0.8, 1]] 
import matplotlib.pyplot as plt
x, y = np.random.multivariate_normal(mean, cov, 500).T
plt.plot(x, y, 'x')
plt.scatter(x,y)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("Histogram of 1x")
plt.xlabel("1x")
plt.ylabel("2x")
plt.xlim(-5,2)
plt.hist(random_numbers[:,0], bins = 50)
plt.show()

plt.title("Histogram of 2x")
plt.xlabel("1x")
plt.ylabel("frequency")
plt.xlim(-5,2)
plt.hist(random_numbers[:,1], bins = 50)
plt.show()
```
